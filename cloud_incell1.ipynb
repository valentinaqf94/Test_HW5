```python
n = 1000
grid = [[[0 for x in xrange(n)] for y in xrange(n)] for z in xrange(n)]
```

```python
m = 1
x = genfromtxt('Serena-Venus.txt',delimiter=" ", usecols=(1))
y = genfromtxt('Serena-Venus.txt',delimiter=" ", usecols=(2))
z = genfromtxt('Serena-Venus.txt',delimiter=" ", usecols=(3))
dltx = (x.max()-x.min())/1000
dlty = (y.max()-y.min())/1000
dltz = (z.max()-z.min())/1000
C = (m/(dltx*dlty*dltz))
```

```python
len(x)
```

```python
for i in xrange(n):
 for j in xrange(n):
  for k in xrange(n):
    if -dltx < (grid[i][j][k]-x[i]):
        wx = (1 + x)/dltx
    elif dltx > (grid[i][j][k]-x[i]):
        wx = (1 - x)/dltx
    else:
        wx = 0
    if -dltx < (grid[i][j][k]-y[i]):
        wy = (1 + y)/dltx
    elif dltx > (grid[i][j][k]-y[i]):
        wy = (1 - y)/dltx
    else:
        wy = 0
    if -dltx < (grid[i][j][k]-z[i]):
        wz = (1 + z)/dltx
    elif dltx > (grid[i][j][k]-z[i]):
        wz = (1 - z)/dltx
    else:
        wz = 0
        
    grid[i][j][k] = C * wx*wy*wz
```
